# Implementação Support Vector Machines (SVM)


## 1) Importar dados 

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

## 2) Carregar dados e devidí-los em teste e treinamento

```python
cancer = datasets.load_breast_cancer()

# Dividi os dados em treino e teste em uma porcetagem de 30% respectivamente 
X_treino, X_teste, y_treino, y_teste = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=109)
```

## 3) Gerar o modelo
Será gerado um modelo com um kernel linear.

```python
# Cria o classificador 
model = svm.SVC(kernel='linear')

# Treina o modelo usando o conjunto de treinamento
model.fit(X_treino, y_treino)

# Prevê a resposta para o conjunto de dados de treino
y_prev = model.predict(X_teste)
```

## 4) Avaliar o modelo
Estimar o quão preciso o modelo é.

```python
# Acurácia: Estimar com que frequência o classificador faz previsões corretas 
print("Acurácia = ", metrics.accuracy_score(y_teste, y_prev))

# Precisão: Com que procentagem as tuplas positivas são classficadas como tais 
print("Precisão = ", metrics.precision_score(y_teste, y_prev))

print("Recall = ", metrics.recall_score(y_teste, y_prev))

cm = confusion_matrix(y_teste, y_prev, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=["Benigno", "Maligno"])
disp.plot()
```
